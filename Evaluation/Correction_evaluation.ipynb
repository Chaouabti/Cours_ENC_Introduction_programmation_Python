### **Exercice 1 : Fonctions et Boucles**

**Sujet :**

Écrivez une fonction nommée `calculer_moyenne` qui :

- Prend en paramètre une **liste de nombres**.
- Renvoie la **moyenne** de ces nombres.
- Si la liste est vide, la fonction doit renvoyer `None`.
- Utilisez une boucle `for` pour parcourir la liste.
- **Commentez** votre code pour expliquer votre démarche.

**Correction :**

````python
def calculer_moyenne(nombres):
    # Vérifier si la liste est vide
    if not nombres:
        return None
    # Initialiser la somme des nombres à zéro
    somme = 0
    # Parcourir chaque nombre dans la liste
    for nombre in nombres:
        somme += nombre  # Ajouter le nombre à la somme
    # Calculer la moyenne en divisant la somme par le nombre d'éléments
    moyenne = somme / len(nombres)
    # Retourner la moyenne calculée
    return moyenne

# Exemple d'utilisation
liste_nombres = [10, 20, 30, 40, 50]
print(calculer_moyenne(liste_nombres))  # Affiche 30.0
````

**Explications :**

- **Vérification de la liste vide :** Avant de procéder au calcul, on vérifie si la liste `nombres` est vide pour éviter une division par zéro.
- **Initialisation de la somme :** On initialise une variable `somme` à zéro pour accumuler les valeurs.
- **Boucle `for` :** On parcourt chaque élément de la liste et on ajoute sa valeur à `somme`.
- **Calcul de la moyenne :** On divise la `somme` totale par le nombre d'éléments dans la liste en utilisant `len(nombres)`.
- **Retour de la moyenne :** La fonction renvoie la valeur calculée de la moyenne.

### **Exercice 2 : Manipulation de Fichiers et du Module `os`**

**Sujet :**

Écrivez un script qui :

1. **Demande** à l'utilisateur de saisir un chemin de répertoire.

2. **Vérifie** si le répertoire existe en utilisant `os.path.exists`.

   - S'il n'existe pas, **le crée** en utilisant `os.makedirs` avec `exist_ok=True`.

3. **Liste** tous les fichiers du répertoire en utilisant `os.listdir` et **affiche** uniquement les fichiers qui ont l'extension `.txt`.

4. Pour chaque fichier .txt, **ouvre** le fichier, **lit** son contenu, et **affiche** le nombre de lignes qu'il contient.

**Correction :**

````python
import os

# 1. Demande à l'utilisateur de saisir un chemin de répertoire
chemin = input("Entrez le chemin du répertoire (laisser vide pour le répertoire courant) : ")

# Utiliser le répertoire courant si aucun chemin n'est fourni
if not chemin:
    chemin = os.getcwd()

# 2. Vérifie si le répertoire existe
if not os.path.exists(chemin):
    # Le crée s'il n'existe pas
    os.makedirs(chemin, exist_ok=True)
    print(f"Le répertoire '{chemin}' a été créé.")
else:
    print(f"Le répertoire '{chemin}' existe déjà.")

# 3. Liste tous les fichiers du répertoire et affiche uniquement les fichiers .txt
    fichiers = os.listdir(chemin)
    fichiers_txt = [f for f in fichiers if f.endswith('.txt')]
    print("Fichiers .txt trouvés :")
    for fichier in fichiers_txt:
        print(fichier)

# 4. Pour chaque fichier .txt, ouvre le fichier, lit son contenu, et affiche le nombre de lignes

for fichier in fichiers_txt:
    chemin_complet = os.path.join(chemin, fichier)
    try:
        lignes = []
        with open(chemin_complet, 'r', encoding='utf-8') as f:
            for line in f:
                lignes.append(line)
            print(f"Le fichier '{fichier}' contient {len(lignes)} lignes.")
    except Exception as e:
        print(f"Impossible de lire le fichier '{fichier}' : {e}")
````

**Explications :**

- **Importation du module `os` :** Nécessaire pour les opérations sur le système de fichiers.
- **Saisie du chemin :** On demande à l'utilisateur d'entrer un chemin ou on utilise le répertoire courant avec `os.getcwd()`.
- **Vérification de l'existence du répertoire :** Utilisation de `os.path.exists()` pour vérifier si le répertoire existe.
- **Création du répertoire :** Si le répertoire n'existe pas, on le crée avec `os.makedirs()` en utilisant `exist_ok=True` pour éviter les erreurs si le répertoire existe déjà.
- **Liste des fichiers :**
  - Utilisation de `os.listdir()` pour obtenir la liste des fichiers.
  - Filtrage des fichiers pour ne garder que ceux qui se terminent par `.txt`.
- **Affichage des fichiers `.txt` :** Impression des noms des fichiers trouvés.
- **Lecture des fichiers :**
  - Initialisation de la liste de ligne pour chaque fichier
  - Pour chaque fichier `.txt`, ouverture en mode lecture.
  - Lecture des lignes avec un boucle `for`.
  - Comptage et affichage du nombre de lignes.
- **Gestion des exceptions :**
  - Utilisation de `try...except` pour gérer les erreurs potentielles lors de la lecture du répertoire ou des fichiers.

### **Exercice 3 : Programmation Orientée Objet**

**Sujet :**

**Création de la classe de base Monument :**

1. Créez une classe nommée Monument avec les attributs suivants :

   - nom (chaîne de caractères)
   - ville (chaîne de caractères)
   - annee_construction (entier)
   - description (chaîne de caractères)

2. La classe doit inclure :

   - Une méthode `caracteristiques` qui affiche les renseignements de base du monument.

**Création de la classe dérivée Chateau :**

3. Créez une classe Chateau qui hérite de la classe Monument et qui ajoute les attributs supplémentaires suivants :

   - style_architectural (chaîne de caractères)
   - classement_unesco (booléen)

4. La classe Chateau doit inclure :

   - Une méthode `caracteristiques` qui affiche les renseignements de la classe parent et les siens.

**Instanciation et utilisation :**

5. Instanciez un objet de la classe Chateau pour le Château de Versailles avec les informations suivantes :

   - Nom : "Château de Versailles"
   - Ville : "Versailles"
   - Année de construction : 1623
   - Description : "Ancienne résidence royale française."
   - Style architectural : "Baroque"
   - Classement UNESCO : True

6. **Appelez** la méthode caracteristiques pour afficher les informations du château.

**Correction :**

````python
# Définition de la classe de base Monument
class Monument:
    def __init__(self, nom, ville, annee_construction, description):
        # Initialisation des attributs
        self.nom = nom
        self.ville = ville
        self.annee_construction = annee_construction
        self.description = description

    def caracteristiques(self):
        # Affichage des informations de base du monument
        print(f"Nom : {self.nom}")
        print(f"Ville : {self.ville}")
        print(f"Année de construction : {self.annee_construction}")
        print(f"Description : {self.description}")

# Classe Chateau qui hérite de Monument
class Chateau(Monument):
    def __init__(self, nom, ville, annee_construction, description, style_architectural, classement_unesco):
        # Appel du constructeur de la classe parent
        super().__init__(nom, ville, annee_construction, description)
        # Initialisation des attributs spécifiques à Chateau
        self.style_architectural = style_architectural
        self.classement_unesco = classement_unesco

    def caracteristiques(self):
        # Appel de la méthode caracteristiques de la classe parent
        super().caracteristiques()
        # Affichage des attributs spécifiques à Chateau
        print(f"Style architectural : {self.style_architectural}")
        print(f"Classé UNESCO : {'Oui' if self.classement_unesco else 'Non'}")
````

````python
# Instanciation du Château de Versailles
chateau_versailles = Chateau(
    nom="Château de Versailles",
    ville="Versailles",
    annee_construction=1623,
    description="Ancienne résidence royale française.",
    style_architectural="Baroque",
    classement_unesco=True
)

# Appel de la méthode caracteristiques
chateau_versailles.caracteristiques()
````

**Explications :**

- **Classe `Monument` :**
  - Le constructeur `__init__` initialise les attributs de base.
  - La méthode `caracteristiques` affiche les informations du monument.
- **Classe `Chateau` :**
  - Hérite de `Monument` et ajoute deux attributs supplémentaires.
  - Le constructeur utilise `super()` pour appeler le constructeur de la classe parent.
  - La méthode `caracteristiques` surcharge celle de la classe parent pour ajouter les nouvelles informations.
- **Instanciation :**
  - Création d'un objet `chateau_versailles` avec les informations fournies.
  - Appel de `caracteristiques` pour afficher toutes les informations du château.
