### ‚öôÔ∏è Exercice 2 ‚Äì √âcrire une fonction bien document√©e (3 pts)

```python
def extrait_mots(texte: str, n: int) -> None:
    """
    Affiche les n premiers mots d'un texte donn√©.

    Param√®tres :
        texte (str) : le texte √† d√©couper
        n (int) : le nombre de mots √† afficher
    Retour :
        None
    """
    # √âtape 1 : s√©parer le texte en mots
    mots = texte.split()
    
    # √âtape 2 : s√©lectionner les n premiers
    extrait = mots[:n]
    
    # √âtape 3 : afficher le r√©sultat
    print(" ".join(extrait))

# Exemple d‚Äôutilisation
extrait_mots("Les humanit√©s num√©riques changent la recherche", 3)
```

### üñºÔ∏è Exercice 3 ‚Äì Les ≈ìuvres d‚Äôart et l‚Äôh√©ritage en Python (4 pts)

```python
class Oeuvre:
    def __init__(self, titre, date, lieu):
        self.titre = titre
        self.date = date
        self.lieu = lieu

    def description(self):
        return f"{self.titre} ‚Äî cr√©√© en {self.date} √† {self.lieu}"


class ArtRenaissance(Oeuvre):
    def __init__(self, titre, date, lieu, courant):
        super().__init__(titre, date, lieu)
        self.courant = courant

    def description_renaissance(self):
        base = super().description()
        return f"{base}. Cette ≈ìuvre s'inscrit dans le courant {self.courant}."


class Sculpture(ArtRenaissance):
    def __init__(self, titre, date, lieu, courant, artiste, pierre, technique):
        super().__init__(titre, date, lieu, courant)
        self.artiste = artiste
        self.pierre = pierre
        self.technique = technique

    def details(self):
        return f"Sculpture de {self.artiste} ‚Äî {self.pierre} ({self.technique})"


s1 = Sculpture(
    "David",
    1504,
    "Florence",
    "Renaissance italienne",
    "Michel-Ange",
    "marbre",
    "taille directe"
)

print(s1.description_renaissance())
print(s1.details())
```
